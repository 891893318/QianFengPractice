```python
user_fare = df_game.loc[:,['名字','支付金额/¥']].groupby(['名字']).sum()
big_r = user_fare[user_fare['支付金额/¥']>6000]
middle_r = user_fare[(user_fare['支付金额/¥']<=6000)&(user_fare['支付金额/¥']>=1000)]
small_r = user_fare[user_fare['支付金额/¥']<100]
df_game.loc[:,['周','支付金额/¥']].groupby('周').sum().sort_values(by='支付金额/¥',ascending=False)
df_game['level'] = 'other'#加一个列 表示大r、中r、小r
#开始标记不同的付费等级
for name_str in big_r.index:
    df_game.loc[df_game['名字'] == name_str,'level']= 'big_r'
for name_str in middle_r.index:
    df_game.loc[df_game['名字'] == name_str,'level']= 'middle_r'
for name_str in small_r.index:
    df_game.loc[df_game['名字'] == name_str,'level']= 'small_r'
#按照类别进行排序
result_level =df_game.loc[:,['周','level','支付金额/¥']].groupby(['周','level']).sum()
def level_display(level_str):
    print("{}按周付费的情况".format(level_str))
    for i in range(13):
        time_str = "第{}周".format(i+1)
        print(time_str,result_level['支付金额/¥'][(time_str,level_str)])
```

```python
level_display('big_r')
level_display('middle_r')
level_display('small_r')
```
